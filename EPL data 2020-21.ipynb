```python
df['minspermatch'] = (df['Mins']/ df['Matches']).astype(int)
df['goalspermatch'] = (df['Goals']/ df['Matches']).astype(float)
df.head()
```

```python
max(df['goalspermatch'])
```

```python
total_goals = df['Goals'].sum()
print(total_goals)
```

```python
total_penalties = df['Penalty_Goals'].sum()
print(total_penalties)
```

```python
total_penaltyattempts = df['Penalty_Attempted'].sum()
print(total_penaltyattempts)
```

```python
penalties_missed = total_penaltyattempts-total_penalties
print(abs(penalties_missed))
```

```python
plt.figure(figsize=(13,6))
data = [total_penalties,penalties_missed]
labels = ['penalty scored','Penalty missed']
color = sns.color_palette('Set2')
plt.pie(data,labels=labels,colors = color,autopct='%.0f%%')
plt.show()
```

```python
assists_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax= sns.barplot(x='Club',y='Assists',data=assists_by_clubs.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists',fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (15,8)
plt.title('plot of clubs vs Total assists',fontsize=20)
```
